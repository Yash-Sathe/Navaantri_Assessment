```python
# CREATE: Insert new record
new_record = {
    'Product': 'New Gadget',        # Replace with your actual columns
    'Category': 'Electronics',
    'Price': 299.99
}
new_df = pd.DataFrame([new_record])
df = pd.concat([df, new_df], ignore_index=True)
print(df.tail())
```

```python
# READ: Retrieve specific records
read_df = df[df['Category'] == 'Books']
print(read_df)
```

```python
#I decided to delete nulls because it very small number compared to number of rows
df.dropna(inplace=True)
```

```python
# UPDATE: Modify existing records
df.loc[df['Product'] == 'Echo Dot', 'Price'] = 39.99
print(df[df['Product'] == 'Echo Dot'])
```

```python
# DELETE: Remove specific records
df = df[df['Category'] != 'Outdated']
print(df)
```

```python
plt.scatter(df['discount_percentage'], df['rating'], color='#009DAC')  
plt.xlabel('Discount Percentage')
plt.ylabel('Rating ')
plt.title('Discount Percentage vs Rating ')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['rating'], bins=20, color='#009DAC', edgecolor='black')
plt.title('Distribution of category Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
